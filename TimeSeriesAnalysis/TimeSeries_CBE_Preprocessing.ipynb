```python
import pandas as pd
import numpy as np
Rainfall_CBE = pd.read_csv ('rainfall_coimbatore.csv')
```

```python
Rainfall_CBE.tail()
```

```python
Rainfall_CBE.columns = ['Year','01','02','03','04','05','06','07','08','09',
                       '10','11','12']
```

```python
Rainfall_CBE.tail()
```

```python
Rain_Transpose = Rainfall_CBE.T
```

```python
Rain_Transpose.head()
```

```python
Rain_Transpose[0][2]
#print(len(Rain_Transpose.columns))
```

```python
Rain_Series = pd.DataFrame()
```

```python
YYYYMM = []
Rainfall = []
for i in range(len(Rain_Transpose.columns)):
    for j in range(1,13):
        if (j<10):
            YearMonth = str(i+1901) + '-0' + str(j)
        else:
            YearMonth = str(i+1901) + '-' + str(j)
        YYYYMM.append (YearMonth)
        Rainfall.append (Rain_Transpose[i][j])
```

```python
YYYYMM[0:5]
```

```python
Rainfall[0:5]
```

```python
Rain_Series['YYYYMM'] = YYYYMM
Rain_Series['Rainfall'] = Rainfall
```

```python
Rain_Series.to_csv ('RainCBE.csv')
```
